Given two below dictionaries, construct the Python program (using functions) to ask users entering their  tastes and then match randomly with the ingredients (print out outputs for user).


```python
question={
    "strong": "Do ye like yer drinks strong?", 
 "salty": "Do ye like it with a salty tang?", 
 "bitter": "Are ye a lubber who likes it bitter?", 
 "sweet": "Would ye like a bit of sweetness with yer poison?", 
 "fruity": "Are ye one for a fruity finish?" }

ingredients = { 
 "strong": ["glug of rum", "slug of whisky", "splash of gin"], 
 "salty": ["olive on a stick", "salt-dusted rim", "rasher of bacon"],  
 "bitter": ["shake of bitters", "splash of tonic", "twist of lemon peel"],  
 "sweet": ["sugar cube", "spoonful of honey", "spash of cola"],  
 "fruity": ["slice of orange", "dash of cassis", "cherry on top"] }
```

# Code

```python
# Import Library
import numpy as np
import pandas as pd
import random
```

```python
#Function
def find_drink():
    for taste in question:
        flavor=str(input(question[taste])).upper()
        if (flavor =="YES") or (flavor=="Y"):
            ans= ingredients[taste]
            return random.choice(ans)
            break
    if (flavor !="YES") or (flavor!="Y"):
        print('Please choose a flavor! and answer YES or NO ')    
        return find_drink()
```
